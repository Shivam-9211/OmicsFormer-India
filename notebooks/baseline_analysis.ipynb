```python
import pandas as pd

from train_random_forest import run_baseline

protein = run_baseline(
    modality="protein",
    h5_file="data/processed/protein_aligned.h5",
    key="protein")

rna = run_baseline(
    modality="rna",
    h5_file="data/processed/rna_aligned.h5",
    key="rna")

mutation = run_baseline(
    modality="mutation",
    h5_file="data/processed/mutation_aligned.h5",
    key="mutation")

methylation = run_baseline(
    modality="methylation",
    h5_file="data/processed/methylation_aligned.h5",
    key="methylation")

df = pd.DataFrame([
    protein,
    rna,
    mutation,
    methylation])

print(df)

df.to_csv(
    "results/table1_baselines.csv",
    index=False)

print("Baseline table saved successfully!")
```
