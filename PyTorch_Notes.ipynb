```python
import torch

# ------------------------------------
# Scalar (0-D)
# ------------------------------------

a = torch.tensor(5)

print("Scalar")
print(a)
print(a.shape)

# ------------------------------------
# Vector (1-D)
# ------------------------------------

b = torch.tensor([1,2,3,4])

print("\nVector")
print(b)
print(b.shape)

# ------------------------------------
# Matrix (2-D)
# ------------------------------------

c = torch.tensor([
    [1,2,3],
    [4,5,6]
])

print("\nMatrix")
print(c)
print(c.shape)

# ------------------------------------
# 3-D Tensor
# ------------------------------------

d = torch.tensor([
    [
        [1,2],
        [3,4]
    ],
    [
        [5,6],
        [7,8]
    ]
])

print("\n3-D Tensor")
print(d)
print(d.shape)
```

```python
import torch

a = torch.tensor([1,2,3])

b = torch.tensor([4,5,6])

print("a =", a)
print("b =", b)

print("\nAddition")
print(a + b)

print("\nSubtraction")
print(a - b)

print("\nMultiplication")
print(a * b)      # element-wise

print("\nDivision")
print(a / b)
```

```python
import torch
import time

device = torch.device("cuda")

x = torch.randn(10000, 10000).to(device)
y = torch.randn(10000, 10000).to(device)

torch.cuda.synchronize()

start = time.time()

z = torch.matmul(x, y)

torch.cuda.synchronize()

print("Time:", time.time() - start)
print("Device:", z.device)
```

```python
import json

with open("data/splits.json", "r") as f:
    splits = json.load(f)

train = set(splits["train"])
val = set(splits["validation"])
test = set(splits["test"])

print("Train size:", len(train))
print("Validation size:", len(val))
print("Test size:", len(test))

print("Train ∩ Validation:", train & val)
print("Train ∩ Test:", train & test)
print("Validation ∩ Test:", val & test)
```
